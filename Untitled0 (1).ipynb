```python
import os
import cv2
import random
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

# 📂 Путь к данным
image_dir = "/content/detection_data/bone fracture detection.v4-v4.yolov8/train/images"
label_dir = "/content/detection_data/bone fracture detection.v4-v4.yolov8/train/labels"

# 🧾 Собираем список пар изображение+аннотация
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
sampled = random.sample(image_files, 5)

# 🖋️ Подгружаем шрифт
try:
  font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 20)
except:
  font = ImageFont.load_default()

for image_file in sampled:
  image_path = os.path.join(image_dir, image_file)
  label_path = os.path.join(label_dir, os.path.splitext(image_file)[0] + ".txt")

  # Загружаем изображение
  img = Image.open(image_path).convert("RGB")
  draw = ImageDraw.Draw(img)

  if os.path.exists(label_path):
    with open(label_path, "r") as f:
      lines = f.readlines()

    if lines:
      for line in lines:
        parts = list(map(float, line.strip().split()))
        if len(parts) == 5:
          cls, x, y, w, h = parts
          W, H = img.size
          x1 = int((x - w/2) * W)
          y1 = int((y - h/2) * H)
          x2 = int((x + w/2) * W)
          y2 = int((y + h/2) * H)
          draw.rectangle([x1, y1, x2, y2], outline="red", width=3)
          draw.text((x1, y1 - 20), "ПЕРЕЛОМ", fill="red", font=font)
    else:
      draw.text((10, 10), "НЕТ ПЕРЕЛОМА", fill="green", font=font)
  else:
    draw.text((10, 10), "НЕТ ПЕРЕЛОМА", fill="green", font=font)

  display(img)
```

```python
def convert_seg_to_bbox(label_dir):
  for filename in os.listdir(label_dir):
    if not filename.endswith(".txt"):
      continue

    file_path = os.path.join(label_dir, filename)

    with open(file_path, "r") as f:
      lines = f.readlines()

    new_lines = []

    for line in lines:
      parts = list(map(float, line.strip().split()))
      if len(parts) <= 5:
        # Уже YOLO формат — не трогаем
        new_lines.append(line)
        continue

      cls = int(parts[0])
      coords = parts[1:]

      xs = coords[0::2]
      ys = coords[1::2]

      x_min = min(xs)
      x_max = max(xs)
      y_min = min(ys)
      y_max = max(ys)

      x_center = (x_min + x_max) / 2
      y_center = (y_min + y_max) / 2
      width = x_max - x_min
      height = y_max - y_min

      new_line = f"{cls} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n"
      new_lines.append(new_line)

    # ⬇️ Вот здесь идёт перезапись старого файла
    with open(file_path, "w") as f:
      f.writelines(new_lines)
convert_seg_to_bbox("/content/detection_data/bone fracture detection.v4-v4.yolov8/train/labels")
convert_seg_to_bbox("/content/detection_data/bone fracture detection.v4-v4.yolov8/valid/labels")
```

```python
def normalize_class_ids(label_dir):
  for filename in os.listdir(label_dir):
    if not filename.endswith(".txt"):
      continue

    path = os.path.join(label_dir, filename)
    with open(path, "r") as f:
      lines = f.readlines()

    new_lines = []
    for line in lines:
      parts = line.strip().split()
      if len(parts) == 5:
        # Заменим class_id на 0
        parts[0] = "0"
        new_lines.append(" ".join(parts) + "\n")
      else:
        print(f"⚠️ Пропущена некорректная строка: {line.strip()}")

    with open(path, "w") as f:
      f.writelines(new_lines)

# Применяем к обеим папкам
normalize_class_ids("/content/detection_data/bone fracture detection.v4-v4.yolov8/train/labels")
normalize_class_ids("/content/detection_data/bone fracture detection.v4-v4.yolov8/valid/labels")
```

```python
from ultralytics import YOLO
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import cv2
from google.colab import files
from IPython.display import display

# ⬇️ Загрузка изображения
uploaded = files.upload()
img_path = list(uploaded.keys())[0]
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 📦 Загрузка моделей
locator = YOLO("runs/detect/train3/weights/epoch49.pt")          # локализатор кости
classifier = YOLO("/content/fracture_detector.pt")                 # классификатор переломов

# 📍 Детекция области кости
results = locator.predict(source=img_rgb, imgsz=640, conf=0.3, verbose=False)[0]

if not results.boxes:
  print("❌ Перелом не найден")
else:
  print(f"🦴 Найдено областей: {len(results.boxes)}")

  # Шрифт для подписи
  try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 32)
  except:
    font = ImageFont.load_default()

  for i, box in enumerate(results.boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0])

    # Расширим bbox немного (на 10% по ширине/высоте)
    pad_x = int((x2 - x1) * 0.1)
    pad_y = int((y2 - y1) * 0.1)
    x1 = max(0, x1 - pad_x)
    y1 = max(0, y1 - pad_y)
    x2 = min(img_rgb.shape[1], x2 + pad_x)
    y2 = min(img_rgb.shape[0], y2 + pad_y)

    # ✂️ Вырезаем область
    crop = img_rgb[y1:y2, x1:x2]

    # 🧠 Классификация перелома
    fracture_result = classifier.predict(source=crop, imgsz=640, conf=0.3, verbose=False)[0]

    # 📋 Подпись
    label = "✅ Перелом найден" if fracture_result.boxes else "❌ Нет перелома"

    # 🔲 Отрисовка
    img_pil = Image.fromarray(img_rgb)
    draw = ImageDraw.Draw(img_pil)
    draw.rectangle([x1, y1, x2, y2], outline="red", width=4)
    draw.text((x1, y1 - 30), label, fill="red", font=font)

    display(img_pil.crop((x1, y1, x2, y2)))
    print(label)

  # Покажем оригинал с прямоугольником
  print("📊 Полный снимок с результатом локализации:")
  display(img_pil)
```
